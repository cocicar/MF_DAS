# **Teмпературски гранични слој на равној плочи**

### **3. Слична решења - произвољна вредност Прантловог броја**


У овом случају, за решење поље брзине се узима Блазијусово решење, у коме су

\begin{equation}
 \eta = y\, \sqrt{\frac{U_{\infty}}{2 \nu x}}, \qquad u = U_{\infty} f(\eta), \quad v = \sqrt{\frac{\nu U_{\infty}}{2x}}\, ( \eta f^{\prime} - f)
 \label{eq:1} \tag{1}
\end{equation}

Дефиницијом бездимензијске температуре $\theta(\eta)$ на исти начин као у претходним случајевима ($T_{\mathrm{w}} = \mathrm{const}$), једначина енергије се своди на једначину

\begin{equation}
\theta(\eta) = \frac{T - T_{\infty}}{T_{\mathrm{w}} - T_{\infty}} \qquad \rightarrow \qquad u \,\frac{\partial T}{\partial x} + v \, \frac{\partial T}{\partial y} = \alpha \,\frac{\partial^2 T}{\partial y^2} \qquad \rightarrow \qquad u \,\frac{\partial \theta}{\partial x} + v \, \frac{\partial \theta}{\partial y} = \alpha \,\frac{\partial^2 \theta}{\partial y^2} 
\label{eq:2}\tag{2}
\end{equation}

Даље је 

\begin{equation}
\frac{\partial \theta}{\partial x} = \frac{\partial \eta}{\partial x}\frac{\mathrm{d}\theta}{\mathrm{d}\eta} = -\frac{1}{2x}\, \eta \, \theta^{\prime}, \qquad \frac{\partial \theta}{\partial y} = \frac{\partial \eta}{\partial y}\frac{\mathrm{d}\theta}{\mathrm{d}\eta} = \sqrt{\frac{U_{\infty}}{2\nu x}}\, \theta^{\prime}, \qquad \frac{\partial^2 \theta}{\partial y^2} = \frac{U_{\infty}}{2 \nu x} \, \theta^{\prime\prime}
\label{eq:3}\tag{3}
\end{equation}

Заменом релација \eqref{eq:1} и \eqref{eq:3} у једначину \eqref{eq:2}, након сређивања се добија **обична диференцијална једначина** са припадајућим граничним условима:

\begin{equation}
  \boxed{\, \theta^{\prime\prime} + \mathrm{Pr}\, f \theta^{\prime} = 0\,} \qquad \eta=0\!: \,\theta = 1 \quad \text{и} \quad \eta \to \infty\!:\, \theta = 0 \label{eq:4}\tag{4}
\end{equation}

где је $\mathrm{Pr} = \nu/\alpha$ Прантлов број. Сада је интересантно направити малу анализу за случај када је вредност Прантловог броја једнака јединици, тј. када је $\mathrm{Pr}=1$. Увођењем нове функције $\vartheta = 1 - \theta$, једначина \eqref{eq:4} и припадајући гранични услови се своде на 

\begin{equation}
  \vartheta^{\prime\prime} + f \vartheta^{\prime} = 0\, \qquad \eta=0\!: \,\vartheta = 0 \quad \text{и} \quad \eta \to \infty\!:\, \vartheta = 1 \label{eq:5}\tag{5}
\end{equation}

Поређењем са једначином за брзински (хидродинамички) гранични слој, као и припадајућим граничним условима (који су идентични)

$$ f^{\prime\prime\prime} + f f^{\prime\prime} = 0 \quad \leftrightarrow \quad (f^{\prime})^{\prime\prime} + f (f^{\prime})^{\prime} = 0\, \qquad \eta=0\!: \,f^{\prime} = 0 \quad \text{и} \quad \eta \to \infty\!:\, f^{\prime} = 1$$

следи да је решење за расподелу бездимензијске температуре за случај када је $\mathrm{Pr} = 1$ директно одређено Блазијусовим решењем, тј. $\vartheta = f^{\prime}$, односно $\theta = 1 - f^{\prime}$.

\begin{equation}
   \boxed{ \, \mathrm{Pr = 1:} \quad \theta(\eta) = 1 - f(\eta)\, }
\end{equation}

У овом случају су дебљине брзинског и температурског граничног слоја једнаке, $\delta = \delta_{th}$. Извод бездимензијске температуре на зиду је једнак другом изводу функције $f$ на зиду, то јест

$$\mathrm{Pr=1\!:} \quad \left(\frac{\mathrm{d}\theta}{\mathrm{d}\eta}\right)_{\!\eta=0} \equiv \left(\frac{\mathrm{d}\theta}{\mathrm{d}\eta}\right)_{\!\mathrm{w}} = -\left(\frac{\mathrm{d}^2 f}{\mathrm{d}\eta^2} \right)_{\!\mathrm{w}} \qquad \boxed{\, \theta^{\prime}_{\mathrm{w}} = - f^{\prime\prime}_{\mathrm{w}} = -0.4696\,}$$

У литератури из области преношења топлоте, овај специјални случај се назива и Рејнолдсова аналогија за случај ламинарног граничног слоја на равној плочи.  

За произвољне вредности Прантловог броја, решићемо нумеричким путем једначину \eqref{eq:4}. У ту сврху увешћемо промењиве $\theta_1(\eta)$ и $\theta_2(\eta)$ и трансформисати обичну диференцијалну једначину другог реда у систем од две обичне диференцијалне једначине првог реда, како следи

\begin{equation}
  \begin{array}{lllll}
     \theta_1 = \theta    &\quad \rightarrow \quad & \theta_1^{\prime} = \theta^{\prime}   & \quad \rightarrow \quad &    \theta_1^{\prime} = \theta_2 \\
     \theta_2 = \theta^{\prime} & \quad \rightarrow \quad & \theta_2^{\prime} = \theta^{\prime\prime} &\quad \rightarrow \quad &  \theta_2^{\prime} = -\mathrm{Pr} f \theta_2
  \end{array} \qquad \qquad \text{Гранични услови:}\quad \theta_1(0) = 1, \,\,\, \theta_1(20) = 0
\end{equation}

```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

# Desna strana sistema diferencijalnih jednacina
def odefun(eta, f):
    f1, f2, f3 = f
    df1 = f2
    df2 = f3
    df3 = -f1*f3
    return [df1, df2, df3]

## Poznati pocetni uslovi
f1_0 = 0   
f2_0 = 0
f3_0 = 0.1

eta_a = 0
eta_b = 20


## SHOOTING - resevanje nelinearne algebarske jednacine Njutn-Rafsonovom metodom
def objective_1(f3_0):
    solution = solve_ivp(odefun, [eta_a, eta_b], [f1_0, f2_0, f3_0[0]], method='DOP853')
    f2 = solution.y[1]
    bc = 1  # гранични услов
    return f2[-1] - bc

f3_0 = fsolve(objective_1, 0.5)[0]
## ------------- END of shooting ------------------------
eta = np.linspace(eta_a, eta_b, 500)
resenje = solve_ivp(odefun, [eta_a, eta_b], [f1_0, f2_0, f3_0], method='DOP853', dense_output=True)
f = resenje.sol(eta)

# Fitovanje resenja za f polinomom 12og stepena 
fit_coeffs = np.polyfit(eta, f[0], 12)
polinom = np.poly1d(fit_coeffs)
# --------------------------------------------

## --------- VREDNOST PRANTLOVOG BROJA ------- ##
Pr = 6
## ------------------------------------------- ##

def theta_fun(eta, theta):
    th1, th2 = theta
    th1 = th2
    th2 = -Pr*polinom(eta)*th2
    return [th1, th2]

th1_0 = 1
def objective_2(th2_0):
    solution = solve_ivp(theta_fun, [eta_a, eta_b], [th1_0, th2_0[0]], method='DOP853')
    th1 = solution.y[0]
    bc_th = 0  # гранични услов
    return th1[-1] - bc_th

th2_0 = fsolve(objective_2, -0.1)[0]
print("Извод бездимензијске температуре на зиду: (theta')_w/sqrt(2) = ", th2_0/np.sqrt(2))
resenje_theta = solve_ivp(theta_fun, [eta_a, eta_b], [th1_0, th2_0], method='DOP853', dense_output=True)
theta = resenje_theta.sol(eta)

plt.figure(figsize=(15, 5))
plt.subplots_adjust(wspace=0.5)
fig1=plt.subplot(121)
plt.title("Профил бездимензијске температуре за $\mathrm{Pr} = %.3f$" % Pr, fontsize=13)
plt.grid(True, linestyle='dashed')
plt.plot(theta[0], eta, lw = 3)
plt.xlabel("$\\theta$", fontsize=12, labelpad=20)
plt.ylabel("$\\eta$", fontsize=12, rotation=0, labelpad=20)

fig1=plt.subplot(122)
plt.title("Поређење профила $\\theta$ и $1 -f^{\prime}$ ($\mathrm{Pr} = %.3f$)" % Pr, fontsize = 13)
plt.grid(True, linestyle='dashed')
plt.plot(theta[0], eta, lw = 2, label='$\\theta$')
plt.plot(1 - f[1], eta, lw = 2, label='$1 - f^{\prime}$')
plt.xlabel("$\\theta, \,\,\, 1-f^{\prime}$", fontsize=12, labelpad=20)
plt.ylabel("$\\eta$", fontsize=12, rotation=0, labelpad=20)
plt.legend()
```

Вредност извода температуре на зиду директно одређује топлотни флукс. Раније је изведена релација која повезује однос Нуселтовог и Рејнолдсовог броја у случају константне температуре плоче. Та релација је 

$$ \frac{\mathrm{Nu}}{\sqrt{\mathrm{Re}}} = -\left(\frac{\partial \theta}{\partial \overline{y}}\right)_{\!\mathrm{w}}, \qquad \mathrm{Nu} = \frac{h(x) L}{\lambda}, \,\,\,\,\mathrm{Re} = \frac{U_{\infty} L}{\nu}; \qquad \overline{y} = \frac{y}{L}\sqrt{\mathrm{Re}}$$ 

Сада ћемо одредити везу између $(\partial \theta/\partial \overline{y})_{\mathrm{w}}$ и $(\mathrm{d} \theta/\mathrm{d} \eta)_{\mathrm{w}}$. У ту сврху ћемо прво трансформисати промењиву сличности

$$\eta =  y\, \sqrt{\frac{U_{\infty}}{2 \nu x}} = \frac{\overline{y} L}{\sqrt{\mathrm{Re}}} \sqrt{\frac{U_{\infty}}{2 \nu x}} = \frac{1}{\sqrt{2 x}} \frac{\overline{y} L}{\sqrt{\mathrm{Re}}} \underbrace{\sqrt{\frac{U_{\infty} L}{ \nu}}}_{\sqrt{\mathrm{Re}}} \frac{1}{\sqrt{L}} = \frac{1}{\sqrt{2x}}\frac{\overline{y} L}{\sqrt{\mathrm{Re}}} \frac{\sqrt{\mathrm{Re}}}{\sqrt{L}} \qquad \Rightarrow \quad \eta = \frac{\overline{y}}{\sqrt{2}} \,\left(\frac{x}{L}\right)^{-1/2}$$

Даље је

$$ \frac{\partial \theta}{\partial \overline{y}} = \frac{\partial \eta}{\partial \overline{y}} \frac{\mathrm{d}\theta}{\mathrm{d}\eta} = \frac{1}{\sqrt{2}} \,\left(\frac{x}{L}\right)^{-1/2} \,\frac{\mathrm{d}\theta}{\mathrm{d}\eta} \qquad \Rightarrow \qquad \left(\frac{\partial \theta}{\partial \overline{y}}\right)_{\!\mathrm{w}} = \frac{\sqrt{2}}{2} \,\left(\frac{x}{L}\right)^{-1/2} \, \left(\frac{\mathrm{d}\theta}{\mathrm{d}\eta}\right)_{\!\mathrm{w}} \qquad \Rightarrow \qquad \frac{\mathrm{Nu}}{\sqrt{\mathrm{Re}}} = \left[-\frac{\sqrt{2}}{2}\,  \theta_{\mathrm{w}}^{\prime}\right] \,\left(\frac{x}{L}\right)^{-1/2}$$

Aко дефинишемо локалне вредности Нуселтовог и Рејнолдсовог броја, где се уместо дужине плоче $L$ користи координата $x$, даље следи

\begin{equation}
\mathrm{Nu}_x = \frac{h(x)\, x}{\lambda}, \quad \mathrm{Re}_x = \frac{U_{\infty}\, x}{\nu} \qquad \Rightarrow \qquad \frac{\mathrm{Nu}}{L} = \frac{\mathrm{Nu}_x}{x}, \quad \frac{\mathrm{Re}}{L} = \frac{\mathrm{Re}_x}{x} \qquad \Rightarrow \qquad \frac{\mathrm{Nu}}{\sqrt{\mathrm{Re}}} = \frac{\mathrm{Nu_x}}{\sqrt{\mathrm{Re_x}}} \,\left(\frac{x}{L}\right)^{-1/2}  \qquad \Rightarrow \qquad \frac{\mathrm{Nu_x}}{\sqrt{\mathrm{Re_x}}} =  -\frac{\sqrt{2}}{2}\,  \theta_{\mathrm{w}}^{\prime}\qquad \Rightarrow \qquad \boxed{\, \mathrm{Nu_x} = \left[-\frac{\sqrt{2}}{2}\,  \theta_{\mathrm{w}}^{\prime}\right] \mathrm{Re_x}^{1/2}\,} \label{eq:6}\tag{6}
\end{equation}

Вредност извода $\theta_{\mathrm{w}}^{\prime}$ зависи од вредности Прантловог броја, јер се он јавља у једначини (4). Корелацију између вредности $\mathrm{Pr}$ и $\theta_{\mathrm{w}}^{\prime}$ je први успоставио Пулхаузен, који је за вредности $\mathrm{Pr} > 0.6$ дошао до следеће корелације

\begin{equation}
  \boxed{\,\mathrm{Nu}_x = 0.332 \, \mathrm{Re}_x^{-1/2}\, \mathrm{Pr}^{1/3}, \quad \text{за}\,\,\,0.6 < \mathrm{Pr} <2000\,}
\end{equation}
